# Tokenize

```python
from nltk.tokenize import word_tokenize
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
word_token = word_tokenize(text)
text_file = open("Output/word_token.txt", "w", encoding='utf-8')
for word in word_token:
    text_file.write(word)
    text_file.write("\n")
text_file.close()
print(len(word_token),"tokens were found with Word Tokenize method.")
```

```python
from nltk.tokenize import WordPunctTokenizer
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
word_punct_token = WordPunctTokenizer().tokenize(text)
text_file = open("Output/word_punct.txt", "w", encoding='utf-8')
for word in word_punct_token:
    text_file.write(word)
    text_file.write("\n")
text_file.close()
print(len(word_punct_token),"tokens were found with Word Punctuation Tokenize method.")
```

```python
from nltk.tokenize import RegexpTokenizer
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
regexp_tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
regexp_token = regexp_tokenizer.tokenize(text)
text_file = open("Output/regexp.txt", "w", encoding='utf-8')
for word in regexp_token:
    text_file.write(word),
    text_file.write("\n")
text_file.close()
print(len(regexp_token),"tokens were found with Regular Expression Tokenize method.")
```

```python
from nltk.tokenize import sent_tokenize
text = open('Input/data.txt', 'r', encoding = 'utf-8')

text_file = open("Output/sent_token.txt", "w", encoding='utf-8')

### This part to remove end line break
string_without_line_breaks = ""
for line in text:
    stripped_line = line.rstrip() + " "
    string_without_line_breaks += stripped_line

#print(string_without_line_breaks)
###
sent_token = sent_tokenize(string_without_line_breaks)
for word in sent_token:
    text_file.write(word)
    text_file.write("\n")

text_file.close()
print(len(sent_token),"sentences were found with Sentence Tokenize method.")
```

# Remove unwanted value from tokenize data

Tokenize data now contain of punctuations, special characters and capital letter and numbers. and some single character, word contain only one character

```python
import re
```

```python
word_punct_token = open("Output/word_punct.txt", "r", encoding='utf-8').readlines()
cleaned_token = open("Output/cleaned_wordpunct_token.txt", "w", encoding='utf-8')
clean_token=[]
for token in word_punct_token:
    new_token = re.sub(r'[^a-zA-Z]+', '', token) # remove any value that are not alphabetical
    if new_token != "" and len(new_token) >= 2: # remove empty value and single character value
        vowels=len([v for v in new_token if v in "aeiou"])
        if vowels != 0: # remove line that only contains consonants
            clean_token.append(new_token)                       
            cleaned_token.write(new_token)
            cleaned_token.write('\n')

cleaned_token.close()
```

```python
word_punct_token = open("Output/word_punct.txt", "r", encoding='utf-8').readlines()
cleaned_token = open("Output/cleaned_wordpunct_token_lowercase.txt", "w", encoding='utf-8')
clean_token=[]
for token in word_punct_token:
    new_token = re.sub(r'[^a-zA-Z]+', '', token) # remove any value that are not alphabetical
    if new_token != "" and len(new_token) >= 2: # remove empty value and single character value
        vowels=len([v for v in new_token if v in "aeiou"])
        if vowels != 0: # remove line that only contains consonants
            new_token = new_token.lower() # change to lower case
            clean_token.append(new_token)                       
            cleaned_token.write(new_token)
            cleaned_token.write('\n')

cleaned_token.close()
```

## Using Stemmer

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
# stemming with cleaned word_punct tokenizer_stopword_removed
input_words = tokens
# Create various stemmer objects
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')

# Create a list of stemmer names for display
stemmer_names = ['PORTER', 'LANCASTER', 'SNOWBALL']
formatted_text = '{:>16}' * (len(stemmer_names) + 1)
text_file = open("Output/stemmed_cleaned_wordpunct_token_removed_stopwords.txt", "w", encoding='utf-8')
text_file.write('\n')
text_file.write(formatted_text.format('INPUT WORD', *stemmer_names))
text_file.write('\n')
text_file.write('='*68)
# Stem each word and display the output
for word in input_words:
    output = [word, porter.stem(word), lancaster.stem(word), snowball.stem(word)]
    text_file.write('\n')
    text_file.write(formatted_text.format(*output))
text_file.close()
```

Comparing Porter, Lancaster and Snowball 3 stemming techniques, Porter is least strict, will Lancaster is the strictest technique, and snowball are moderate, hence **Snowball will be used in this project.**

```python
# extract only snowball input
from nltk.stem.snowball import SnowballStemmer
# stemming with cleaned_wordpunct_token_removed_stopwords
input_words = tokens
# Create various stemmer objects
snowball = SnowballStemmer('english')
text_file = open("Output/Snowball_cleaned_wordpunct_token_removed_stopwords.txt", "w", encoding='utf-8')
# Stem each word and display the output
for word in input_words:
    output = snowball.stem(word)
    text_file.write(output)
    text_file.write('\n')
text_file.close()
```

## Extract Text

```python
## read sentences and extract only line which contain the keywords
import pandas as pd
import re
# open file
prison_keyword = open('prison keyword.txt', 'r', encoding = 'utf-8').readlines()
texts = open('Output/sent_token.txt', 'r', encoding = 'utf-8').readlines()
# define function to read file and remove next line symbol
def read_file(file):
    texts = []
    for word in file:
        text = word.rstrip('\n')
        texts.append(text)

    return texts

# save to variable        
key = read_file(prison_keyword)
corpus = read_file(texts)

# open file to write line which contain keywords
file = open('Output/keyline_prison.txt', 'w', encoding = 'utf-8') 
def write_file(file, keyword, corpus):    
    keyline = []      
    for line in corpus:
        line = line.lower()
        for key in keyword:
            result = re.search(r"(^|[^a-z])" + key + r"([^a-z]|$)", line)
            if result != None:
                keypair = [key, line]
                keyline.append(keypair) 
                file.write(line)                                
                break    # to prevent copy the same line from diferent keyword to lower file size                
            else:
                pass                                            
          
    return(keyline)

output = write_file(file,key,corpus)
# create DataFrame using data 
df = pd.DataFrame(output, columns =['Key', 'Line']) 
# save to csv file    
df.to_csv("Output/df_keyline_prison.csv")
write_file(file,key,corpus)
print(df.shape)
```

```python
x = range(0, 50)
plt.plot(x, coherence_scores)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.title("Bigram LDA")
plt.show()
```

# LDA Gensim Model

```python
import re
import numpy as np
import pandas as pd
from pprint import pprint
import nltk
# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

# spacy for lemmatization
import spacy

# Plotting tools
import pyLDAvis
import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm
# Enable logging for gensim - optional
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.ERROR)

import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
texts = open('Output/sent_token.txt', 'r', encoding = 'utf-8').readlines()
```

```python
texts[0]
```

```python
data = [re.sub('\s+',' ', sent) for sent in texts] #remove new line character
data = [re.sub("\''","", sent) for sent in texts] # remove single quotes
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(data))

print(data_words[:2])
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[data_words[0]]])
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[data_words[0]]])
```

```python
# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
# Remove Stop Words
data_words_nostops = remove_stopwords(data_words)

# Form Bigrams
data_words_bigrams = make_bigrams(data_words_nostops)

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
# python3 -m spacy download en
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

# Do lemmatization keeping only noun, adj
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ'])

print(data_lemmatized[:1])
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1])
```

```python
id2word[0]
```

```python
# Human readable format of corpus (term-frequency)
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

```python
# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=20, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
