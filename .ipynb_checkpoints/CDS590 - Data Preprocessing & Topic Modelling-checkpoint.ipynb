# Tokenize

```python
from nltk.tokenize import word_tokenize
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
word_token = word_tokenize(text)
text_file = open("Output/word_token.txt", "w", encoding='utf-8')
for word in word_token:
    text_file.write(word)
    text_file.write("\n")
text_file.close()
print(len(word_token),"tokens were found with Word Tokenize method.")
```

```python
from nltk.tokenize import WordPunctTokenizer
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
word_punct_token = WordPunctTokenizer().tokenize(text)
text_file = open("Output/word_punct.txt", "w", encoding='utf-8')
for word in word_punct_token:
    text_file.write(word)
    text_file.write("\n")
text_file.close()
print(len(word_punct_token),"tokens were found with Word Punctuation Tokenize method.")
```

```python
from nltk.tokenize import RegexpTokenizer
text = open('Input/data.txt', 'r', encoding = 'utf-8').read().replace("\n\n\x0c","")
regexp_tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
regexp_token = regexp_tokenizer.tokenize(text)
text_file = open("Output/regexp.txt", "w", encoding='utf-8')
for word in regexp_token:
    text_file.write(word),
    text_file.write("\n")
text_file.close()
print(len(regexp_token),"tokens were found with Regular Expression Tokenize method.")
```

```python
from nltk.tokenize import sent_tokenize
text = open('Input/data.txt', 'r', encoding = 'utf-8')

text_file = open("Output/sent_token.txt", "w", encoding='utf-8')

### This part to remove end line break
string_without_line_breaks = ""
for line in text:
    stripped_line = line.rstrip() + " "
    string_without_line_breaks += stripped_line

#print(string_without_line_breaks)
###
sent_token = sent_tokenize(string_without_line_breaks)
for word in sent_token:
    text_file.write(word)
    text_file.write("\n")

text_file.close()
print(len(sent_token),"sentences were found with Sentence Tokenize method.")
```

# Remove unwanted value from tokenize data

Tokenize data now contain of punctuations, special characters and capital letter and numbers. and some single character, word contain only one character

```python
import re
```

```python
word_punct_token = open("Output/word_punct.txt", "r", encoding='utf-8').readlines()
cleaned_token = open("Output/cleaned_wordpunct_token.txt", "w", encoding='utf-8')
clean_token=[]
for token in word_punct_token:
    new_token = re.sub(r'[^a-zA-Z]+', '', token) # remove any value that are not alphabetical
    if new_token != "" and len(new_token) >= 2: # remove empty value and single character value
        vowels=len([v for v in new_token if v in "aeiou"])
        if vowels != 0: # remove line that only contains consonants
            clean_token.append(new_token)                       
            cleaned_token.write(new_token)
            cleaned_token.write('\n')

cleaned_token.close()
```

```python
word_punct_token = open("Output/word_punct.txt", "r", encoding='utf-8').readlines()
cleaned_token = open("Output/cleaned_wordpunct_token_lowercase.txt", "w", encoding='utf-8')
clean_token=[]
for token in word_punct_token:
    new_token = re.sub(r'[^a-zA-Z]+', '', token) # remove any value that are not alphabetical
    if new_token != "" and len(new_token) >= 2: # remove empty value and single character value
        vowels=len([v for v in new_token if v in "aeiou"])
        if vowels != 0: # remove line that only contains consonants
            new_token = new_token.lower() # change to lower case
            clean_token.append(new_token)                       
            cleaned_token.write(new_token)
            cleaned_token.write('\n')

cleaned_token.close()
```

## Using Lemmatizer

As nouns and adjectives containing more information, we will only select noun and verb to lemmatize it.

```python
import nltk
from nltk.stem import WordNetLemmatizer
# open the POS Tagged text
tagset = pd.read_csv("Output/data_tagset.csv")
tagset_allowed = ['NN','NNS','NNP','NNPS','JJ','JJR','JJS']
new_tagset = tagset.loc[tagset['Tag'].isin(tagset_allowed)]
text = [str(x) for x in new_tagset['Word']]
tag =[x for x in new_tagset['Tag'] if x != '']         
# Create lemmatizer object 
lemmatizer = WordNetLemmatizer()

# Lemmatize each word and display the output
lemmatize_text = []
for word in text:
    output = [word, lemmatizer.lemmatize(word, pos='n'),lemmatizer.lemmatize(word, pos='a')]
    lemmatize_text.append(output)

# create DataFrame using data 
df = pd.DataFrame(lemmatize_text, columns =['Word', 'Lemmatized Noun', 'Lemmatized Adjective']) 
#print(df.head(50))
df['Tag']=tag
#print(df.head(50))
# save to csv file    
df.to_csv("Output/df_lemmatized_noun_adjectives.csv")   
```

## Using Stemmer

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
# stemming with cleaned word_punct tokenizer_stopword_removed
input_words = tokens
# Create various stemmer objects
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')

# Create a list of stemmer names for display
stemmer_names = ['PORTER', 'LANCASTER', 'SNOWBALL']
formatted_text = '{:>16}' * (len(stemmer_names) + 1)
text_file = open("Output/stemmed_cleaned_wordpunct_token_removed_stopwords.txt", "w", encoding='utf-8')
text_file.write('\n')
text_file.write(formatted_text.format('INPUT WORD', *stemmer_names))
text_file.write('\n')
text_file.write('='*68)
# Stem each word and display the output
for word in input_words:
    output = [word, porter.stem(word), lancaster.stem(word), snowball.stem(word)]
    text_file.write('\n')
    text_file.write(formatted_text.format(*output))
text_file.close()
```

Comparing Porter, Lancaster and Snowball 3 stemming techniques, Porter is least strict, will Lancaster is the strictest technique, and snowball are moderate, hence **Snowball will be used in this project.**

```python
# extract only snowball input
from nltk.stem.snowball import SnowballStemmer
# stemming with cleaned_wordpunct_token_removed_stopwords
input_words = tokens
# Create various stemmer objects
snowball = SnowballStemmer('english')
text_file = open("Output/Snowball_cleaned_wordpunct_token_removed_stopwords.txt", "w", encoding='utf-8')
# Stem each word and display the output
for word in input_words:
    output = snowball.stem(word)
    text_file.write(output)
    text_file.write('\n')
text_file.close()
```

# WordCloud on the Text Extracted based on the Keywords

## Extract Text

```python
## read sentences and extract only line which contain the keywords
import pandas as pd
import re
# open file
prison_keyword = open('prison keyword.txt', 'r', encoding = 'utf-8').readlines()
texts = open('Output/sent_token.txt', 'r', encoding = 'utf-8').readlines()
# define function to read file and remove next line symbol
def read_file(file):
    texts = []
    for word in file:
        text = word.rstrip('\n')
        texts.append(text)

    return texts

# save to variable        
key = read_file(prison_keyword)
corpus = read_file(texts)

# open file to write line which contain keywords
file = open('Output/keyline_prison.txt', 'w', encoding = 'utf-8') 
def write_file(file, keyword, corpus):    
    keyline = []      
    for line in corpus:
        line = line.lower()
        for key in keyword:
            result = re.search(r"(^|[^a-z])" + key + r"([^a-z]|$)", line)
            if result != None:
                keypair = [key, line]
                keyline.append(keypair) 
                file.write(line)                                
                break    # to prevent copy the same line from diferent keyword to lower file size                
            else:
                pass                                            
          
    return(keyline)

output = write_file(file,key,corpus)
# create DataFrame using data 
df = pd.DataFrame(output, columns =['Key', 'Line']) 
# save to csv file    
df.to_csv("Output/df_keyline_prison.csv")
write_file(file,key,corpus)
print(df.shape)
```

# Plot Frequency Plot of the Words

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
def plot_10_most_common_words(count_data, count_vectorizer):
    words = count_vectorizer.get_feature_names()
    total_counts = np.zeros(len(words))
    for t in count_data:
        total_counts+=t.toarray()[0]
    
    count_dict = (zip(words, total_counts))
    count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:10]
    words = [w[0] for w in count_dict]
    counts = [w[1] for w in count_dict]
    x_pos = np.arange(len(words)) 
    
    plt.figure(2, figsize=(18, 15/1.6180))
    plt.subplot(title='10 Most Common Words')
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2.5})
    sns.barplot(x_pos, counts, palette='husl')
    plt.xticks(x_pos, words, rotation=90) 
    plt.xlabel('Words')
    plt.ylabel('Counts')
    plt.savefig('Vis/10_most_common_words.png')
    plt.show()# Initialise the count vectorizer with the English stop words
count_vectorizer = CountVectorizer()# Fit and transform the processed titles
count_data = count_vectorizer.fit_transform(df_lemmatized['Lemmatized Word'])# Visualise the 10 most common words
plot_10_most_common_words(count_data, count_vectorizer)
```

# WordCloud Bigram

**ngram_range**  tuple (min_n, max_n), default=(1, 1)
The lower and upper boundary of the range of n-values for different word n-grams or char n-grams to be extracted. All values of n such such that min_n <= n <= max_n will be used. For example an ngram_range of (1, 1) means only unigrams, (1, 2) means unigrams and bigrams, and (2, 2) means only bigrams. Only applies if analyzer is not callable. [Documentation Refer Here](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)

```python
#Using count vectoriser to view the frequency of bigrams
lemmatized_word = ' '.join(str(x) for x in df_lemmatized['Lemmatized Word'])
lemmatized_word = [lemmatized_word]
vectorizer = CountVectorizer(ngram_range=(2, 2))
bag_of_words = vectorizer.fit_transform(lemmatized_word)
vectorizer.vocabulary_
sum_words = bag_of_words.sum(axis=0) 
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
print (words_freq[:100])
```

```python
#Generating wordcloud and saving as jpg image
words_dict = dict(words_freq)
WC_height = 800
WC_width = 1600
WC_max_words = 200
wordCloud = WordCloud(max_words=WC_max_words, height=WC_height, width=WC_width,background_color = 'white')
wordCloud.generate_from_frequencies(words_dict)
plt.title('Most frequently occurring bigrams')
plt.imshow(wordCloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordCloud.to_file('vis/wordcloud_bigram.jpg')
```

# Topic Modelling using nltk.ngrams

```python
# import libraries & define n_grams function
import re
import numpy as np
import pandas as pd
from pprint import pprint
import nltk
from nltk.util import ngrams

# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

# import algorithm of topic modelling, LDA down not need to be import as libraries
from gensim.models import HdpModel
from gensim.models import LsiModel
from gensim.models import nmf

# Plotting tools
import pyLDAvis
import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm
# Enable logging for gensim - optional
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.ERROR)

import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)


# run this line if previous code is not run yet
df_lemmatized = pd.read_csv('Output/df_model_input.csv')

# Function to generate n-grams from sentences.
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [' '.join(grams) for grams in n_grams]

lemmatized_word = ' '.join(str(x) for x in df_lemmatized['Lemmatized Word'])
```

## 1. Topic Modelling - Create Corpus and Dictionary (Bigram)

```python
# create bigrams
data_words_bigrams = extract_ngrams(lemmatized_word,2)
print("Random Bigram Example",data_words_bigrams[1234:1242])

# Create Dictionary
text_file = [bigram.split() for bigram in data_words_bigrams]
id2word = corpora.Dictionary(text_file)

# Create Corpus
texts = text_file

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print("Corpus: ",corpus[:1])

# Human readable format of corpus (term-frequency)
print("Corpus with original text: ",[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]])
```

### 1.1 LDA

```python
# Loop LDA  model with different number of topics (test from 1 to 50)
i=1
coherence_scores=[]
while i<50:
    # Build LDA model
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=id2word,num_topics=i)
    LDA_Topics = lda_model.print_topics()
    doc_lda = lda_model[corpus]
    
    # Compute Coherence Score
    coherence_model_lda = CoherenceModel(model=lda_model, corpus=corpus, dictionary=id2word, coherence='u_mass')
    coherence_lda = coherence_model_lda.get_coherence()
    score = [i,coherence_lda]
    coherence_scores.append(score) 
    i+=1
```

```python
# create DataFrame using score and retrieve total number of topics with minimum score
df = pd.DataFrame(coherence_scores, columns =['Total Number of Topics', 'Coherence Score']) 
print("The total number of topics with the minimum coherence score are as follow:")
df.loc[(df['Coherence Score'] == df['Coherence Score'].min())]
```

```python
i = df['Total Number of Topics']
score = df['Coherence Score']
plt.plot(i, score)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.title("Bigram LDA")
plt.show()
```

```python
# save output of best model
output = open("Model Output/lda_best_output_bigram.txt",'w')
# Build LDA model
# find the total number of topics with minimum coherence score
#i = df.loc[df['Coherence Score'] == df['Coherence Score'].min(),'Total Number of Topics'].iloc[0]

lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=id2word,num_topics=20)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
LDA_Topics = lda_model.print_topics()
for topic in LDA_Topics:
    output.write(' '.join(str(item) for item in topic) + '\n')

doc_lda = lda_model[corpus]
    
# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, corpus=corpus, dictionary=id2word, coherence='u_mass')
coherence_lda = coherence_model_lda.get_coherence()
#print('Coherence Score: ', coherence_lda)
output.write('Coherence Score: ')
output.write(str(coherence_lda)+'\n')   
output.close()
```
